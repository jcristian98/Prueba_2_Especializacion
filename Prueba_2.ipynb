## **Ejercicio 3 (30%)**

Importe el dataset de nba del repositorio de clase y responda (hint: Elimine las instancias con registros faltantes):

1. Cuáles son los equipos que pagan los salarios más altos (presente el resultado como un dataFrame con el nombre del equipo y la cantidad de dinero de la nómina)
2. Separe la variable *Age* en tres grupos y cree una nueva columna llamada *Age-Group* con esta información.<
3. Construya una tabla dinámica que muestre el salario promedio de los jugadores por equipo y posición para para los cinco equipos que pagan mejores salarios.

```python
# Importa la biblioteca pandas
# Lee el archivo CSV desde la URL proporcionada y carga los datos en un DataFrame
import pandas as pd
nba=pd.read_csv("https://raw.githubusercontent.com/tomasate/Datos_Clases/main/Datos_1/nba.csv",index_col=0)
nba.head()

# Proporciona un resumen útil de la estructura y el contenido del DataFrame
nba.info()

# Elimina las instancias con registros faltantes
nba_cleaned = nba.dropna()
nba_cleaned.info()
```

```python
# Agrupamos por equipo y sumamos los salarios de cada equipo
df=nba_cleaned.groupby('Team',as_index=False)[['Salary']].sum()
df=df.sort_values(by='Salary', ascending=False)
df.reset_index(drop=True, inplace=True)
df=df.head()
df
```

```python
# Proporciona un resumen estadístico útil que describe la distribución de los valores en una Serie.
# Para poder mirar en que rango se encuentra la variable Age
nba_cleaned.Age.describe()

# Definimos los intervalos de los grupos de edades
bins = [nba_cleaned.Age.min(), 24, 29, nba_cleaned.Age.max()]

# Definimos las etiquetas para los grupos de edades
labels = ['Young', 'Middle-aged', 'Senior']

# Segmentamos las edades en grupos y asignamos etiquetas
nba_cleaned['Age-Group'] = pd.cut(nba_cleaned.Age, bins=bins, labels=labels)
nba_cleaned
```

```python
# Filtramos el DataFrame original para incluir solo los jugadores de los cinco equipos seleccionados
nba_top_5_teams = nba_cleaned[nba_cleaned['Team'].isin(df['Team'])]

# Creamos la tabla dinámica
tabla_dinamica = pd.pivot_table(nba_top_5_teams, index='Team', columns='Position', values='Salary', aggfunc='mean')
tabla_dinamica
```
